```python
import numpy as np
import seaborn as sns
```

```python
# import datasets
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
sns.stripplot(data=tips,x='day',y='total_bill')
```

```python
sns.stripplot(data=tips,x='day',y='total_bill',jitter=False)
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='strip')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='strip',hue='sex',jitter=0.3)
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='swarm')
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
sns.boxplot(data=tips,y='total_bill')
```

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',split=True)
```

```python
sns.barplot(data=tips,x='sex',y='total_bill')
```

```python
sns.barplot(data=tips,x='sex',y='total_bill',hue='smoker',estimator=np.min)
```

```python
sns.pointplot(data=tips,x='sex',y='total_bill',hue='smoker',estimator=np.mean)
```

```python
sns.countplot(data=tips,x='sex',hue='day')
```

```python
sns.catplot(data=tips,x='sex',y='total_bill',col='smoker',kind='box',row='time')
```

```python
sns.regplot(data=tips,y='tip',x='total_bill')
```

```python
sns.lmplot(data=tips,y='tip',x='total_bill',hue='sex')
```

```python
sns.residplot(data=tips,y='tip',x='total_bill')
```

```python
sns.catplot(data=tips,x='sex',y='total_bill',kind='violin',col='day')
```

```python
g = sns.FacetGrid(data=tips,col='day',row='time')
g.map(sns.violinplot,'sex','total_bill')
```

```python
sns.pairplot(iris,hue='species')
```

```python
g = sns.PairGrid(data=iris,hue='species')
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```python
g = sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```python
sns.jointplot(data=tips,y='tip',x='total_bill')
```

```python
g = sns.JointGrid(data=tips,y='tip',x='total_bill')
g.plot(sns.scatterplot,sns.histplot)
```
